```python
import csv

with open('data.csv', newline='') as f:
    reader = csv.reader(f)
    data_temp = list(reader)
data_temp.pop(0)
print(data_temp[:10])
```

```python
import matplotlib.pyplot as plt
x = [float(x[3]) for x in data_temp]
xline = [0.1 for _ in data_temp]
plt.plot(x)
plt.plot(xline)
plt.show()
```

```python
data_new = []
for x in data_temp:
    if float(x[3]) > 0.1:
        x[6] = 'True'
    else:
        x[6] = 'False'
    data_new.append(x)
```

```python
import csv

f = open(f"data_new.csv", 'w', newline='', encoding='utf-8')
writer = csv.writer(f)
writer.writerows(data_new)
f.close()
```
