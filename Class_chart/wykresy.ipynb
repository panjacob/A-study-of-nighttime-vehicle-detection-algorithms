```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["Jest pojazd", "Nie ma pojazdu"]

data = [43_335 + 10_340, 39_425 + 9_290]
ingredients = [x for x in recipe]


def func(pct, allvals):
    absolute = int(np.round(pct / 100. * np.sum(allvals)))
    return "{:.1f}%\n({:d} obrazów)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Klasy",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

# ax.set_title("Dataset classes count")

plt.show()
```

```python
def f1(precision=0.46, recall=0.95):
    f1 = (2 * precision * recall) / (precision + recall)
    return round(f1, 2)


text = """        Prosty algorytm & 0,85 & 0,91 & 0,82 \\ \hline
        Autorska sieć splotowa 224  & 0,87 & 0,98 & 0,75\\ \hline""".replace('\\ \hline', '')

for x in text.replace(',', '.').split('\n'):
    y = x.split('&')
    # print(y)
    precision = float(y[2])
    recall = float(y[3])
    name = y[0]
    # print(recall)
    print(name, f1(precision, recall))

```
